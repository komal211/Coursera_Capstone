```python
import requests # library to handle requests
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
import random # library for random number generation

#!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim # module to convert an address into latitude and longitude values

# libraries for displaying images
from IPython.display import Image 
from IPython.core.display import HTML 


from IPython.display import display_html
import pandas as pd
import numpy as np
    
# tranforming json file into a pandas dataframe library
from pandas.io.json import json_normalize

!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

print('Folium installed')
print('Libraries imported.')
```

```python

source_of_data = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup_obj=BeautifulSoup(source_of_data,'lxml')
print(soup_obj.title)
from IPython.display import display_html
table_obj = str(soup_obj.table)
display_html(table_obj,raw=True)
```

```python
df1=pd.read_html(table_obj)
df=df1[0]
df.head()
```

```python
#Data preprocessing and cleaning
```

```python
df2=df[df.Borough != 'Not assigned']
#drop the rows where Borough is not assigned

df3=df2.groupby(['Postal Code','Borough'],sort=False).agg(','.join)
df3.reset_index(inplace=True)

df3['Neighborhood']=np.where(df2['Neighborhood'] == 'Not assigned', df3['Borough'],df3['Neighborhood'])
df3
```

```python
df3.shape
```

```python
#part2
#Import a csv file that has the geographical coordinates of each postal code:
lat_and_long = pd.read_csv('https://cocl.us/Geospatial_data')
lat_and_long.head()
```

```python
df4=pd.merge(df3,lat_and_long,on='Postal Code')
df4.head()
```

```python
df5=df4[df4['Borough'].str.contains('Toronto',regex=False)]
df5
```

```python
#part 3 K means clustering
```

```python
tor_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,longt,borough,neighborhood in zip(df5['Latitude'],df5['Longitude'],df5['Borough'],df5['Neighborhood']):
    label = '{}, {}'.format(neighborhood,borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,longt],radius=5,popup=label,color='yellow',fill=True,fill_color='D7BDE2',fill_opacity=0.7,parse_html=False).add_to(tor_map)
        
tor_map
```

```python
k=5
tor_clust=df5.drop(['Postal Code','Borough','Neighborhood'],1)
k_clust = KMeans(n_clusters = k,random_state=0).fit(tor_clust)
k_clust.labels_
df5.insert(0,'Cluster Labels',k_clust.labels_)
```

```python

df5
```

```python
clust_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

x_col = np.arange(k)
y_col = [i + x_col + (i*x_col)**2 for i in range(k)]
arr_color=cm.rainbow(np.linspace(0,1,len(y_col)))
rainbow=[colors.rgb2hex(i) for i in arr_color]

col_mark = []
for lat,longt,neighborhood,cluster in zip(df5['Latitude'],df5['Longitude'],df5['Neighbourhood'],
                                         df5['Cluster Labels']):
    label=folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker([lat,longt],radius=5,popup=label,color=rainbow[cluster-1],fill=True,fill_color=rainbow[cluster-1],fill_opacity=0.6).add_to(clust_map)

clust_map
```
